## RSNA Image Data를 이용해서 3000개의 이미지 파일 추려내기
### class0.csv: class 0 patiendID 1500, class1.csv: class 1  patiendID 1500
#### csv 파일의 patiendID를 가져와서 해당하는 image만 다른 폴더에 CP하기

```python
import numpy as np
import pandas as pd
import shutil
import os
```

```python
classfile = pd.read_csv('./stage_1_train3000_labels.csv')
```

```python
# check data frame
classfile.head()
```

```python
original_image_dir = "./stage_1_train_images/"
move_image_dir = "./train_images/"
image_count = 0
image_nocount=0
for row in classfile['patientId']:
    src = original_image_dir + row + ".dcm"
    des = move_image_dir + row + ".dcm"    
    if os.path.exists(src):
        shutil.move(src, des)
        image_count += 1
    else :
        image_nocount+=1
        
print(image_count)
print(image_nocount)
```

```python
original_image_dir = "./stage_1_train_images/"
move_image_dir = "./train_images/"
count = 0
nocount = 0
for row in classfile['patientId']:
    des = move_image_dir + row + ".dcm"
    if os.path.exists(des):
        count += 1
    else :
        nocount += 1
        
print(count)
print(nocount)
```
